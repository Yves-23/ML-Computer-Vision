```python
dataset = pd.read_excel('breast_cancerdataset1.xlsx')
```

```python
print(dataset.shape)
print(dataset.head(5))
```

```python
X = dataset.iloc[:, :-1].values
X
y= dataset.iloc[:, -1].values
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler


sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
#print(sc.n_features_in_)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(random_state=0)
model.fit(X_train, y_train)
```

```python
Clump_Thickness = int(input("Clump_Thickness: "))
Uniformit_y_of_Cell_Size = int(input("Uniformit_y_of_Cell_Size: "))
Uniformit_y_of_Cell_Shape = int(input("Uniformit_y_of_Cell_Shape: "))
Marginal_Adhesion = int(input("Marginal_Adhesion: "))
Single_Epithelia_I_Cell_Size = int(input("Single_Epithelia_I_Cell_Size: "))
Bare_Nucle_i = int(input("Bare_Nucle_i: "))
Bland_Chromatin = int(input("Bland_Chromatin: "))
Normal_Nucleol_i = int(input("Normal_Nucleol_i: "))
Mitoses = int(input("Mitoses: "))
#Class = int(input("Class: "))

Pred_Values = [[Clump_Thickness, Uniformit_y_of_Cell_Size, Uniformit_y_of_Cell_Shape, Marginal_Adhesion, Single_Epithelia_I_Cell_Size,
                Bare_Nucle_i, Bland_Chromatin , Normal_Nucleol_i , Mitoses]]
result = model.predict(sc.transform(Pred_Values))
#scaled_Pred_Values = sc.transform(Pred_Values)

#result = model.predict(sc.transform(Pred_Values))
#result = model.predict(scaled_Pred_Values)
print(result)
if result == 2:
  print("The patient has benign cancer")
elif result == 4:
  print("The patient has malignant cancer")
else:
  print("no cancer")
```

```python
y_pred = model.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
```

```python
print("Accuracy of the model: {0}%".format(accuracy_score(y_test, y_pred)*100))
```
